```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ptick

# 出力を16:9に設定
plt.figure(figsize=(16, 9))

# フォント設定
plt.rcParams['font.family'] = 'IPAexGothic'
plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['font.size'] = 22

# 軸の設定
#* 軸と文字の距離（x軸とy軸の0がぶつからないようにするため）
plt.rcParams['xtick.major.pad'] = 8
plt.rcParams['ytick.major.pad'] = 8
#* 内向き目盛
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
#* 目盛サイズ
plt.tick_params(which='major', width = 2, length = 10)
plt.tick_params(which='minor', width = 1.5, length = 10)

# グレイスケール
plt.style.use('grayscale')

# x, y
x = np.arange(0.001, 10000, 0.01)
y = np.sin(x)+2*np.log10(x)

# グラフの描画
plt.plot(x, y, c='black')

# 描画範囲の設定
xlimit = 5 * np.pi
plt.xlim([0, xlimit])

# 目盛の範囲設定
plt.xticks(np.arange(0, xlimit+2.1, 2))

# 対数目盛
# plt.xscale('log')

# 現在のグラフを取得して目盛表示を小数表示に変更
ax = plt.gca()
ax.xaxis.set_major_formatter(ptick.FormatStrFormatter('%.2f'))

# ラベルの設定
plt.xlabel('$x$軸')
plt.ylabel('$y$軸')

# グラフの表示
plt.show()

# グラフの保存
#plt.savefig("graph.pdf")
```
